```python
# Load data from sample csv
# data_url = 'https://raw.githubusercontent.com/mapbox/mapboxgl-jupyter/master/examples/points.csv'
# df = pd.read_csv(data_url)
file_name='restaurant_drop_duplicates_in_Austin_TX.csv'
df_res=pd.read_csv(file_name)
df_res=df_res.fillna(0)
df_res.head()
```

```python
df_for_merge=df_res.groupby(['Zip','Price_Level(Google)'])['Address'].count().unstack(level='Price_Level(Google)')
df_for_merge=df_for_merge.fillna(0).reset_index()
df_for_merge
```

```python
zip_geo=pd.read_csv('US Zip Codes from 2013 Government Data')
zip_geo.columns=['Zip','LAT','LNG']
zip_merge=pd.merge(zip_geo,df_for_merge,on=['Zip'],how='inner')
zip_merge['Total']=zip_merge[zip_merge.columns[3:]].sum(axis=1)
zip_merge['High End']=zip_merge[3.0]+zip_merge[4.0]
zip_merge.columns=['Zip', 'LAT', 'LNG', 'NA', "$", "$$", '$$$', "$$$$", 'Total', 'High End']
zip_merge.columns
```
