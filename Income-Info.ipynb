```python
import pandas as pd
```

```python
income_data = pd.read_csv("median_zip.csv")
income_data.head()
```

```python
income_data.dtypes
```

```python
austin_zips = pd.read_csv("austinzip.csv", header=None)
austin_zips.head()
```

```python
austin_zips.rename(columns={0:'Zip'},inplace=False)
austin_zips.head()
```

```python
# for index, row in income_data.iterrows():
#     row['Median'] = str(row['Median']).replace(',','')
new_median=[i.replace(',','') for i in income_data['Median'].values.tolist()]
new_median_num=[int(i) for i in new_median]
new_population = [i.replace(',','') for i in income_data['Pop'].values.tolist()]
new_population_num=[int(i) for i in new_population]

income_no_commas = pd.DataFrame({})
income_no_commas["Zip"] = income_data["Zip"]
income_no_commas["Median"] = new_median_num
income_no_commas["Population"] = new_population_num

income_no_commas.head()
```

```python
austin_income_df = pd.merge(austin_zips, income_no_commas, on="Zip", how="left")
austin_income_df.to_csv("Austin_median_income_by_zip.csv")
austin_income_df
```
